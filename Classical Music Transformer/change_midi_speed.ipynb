```python
# TODO
midi_path = '/gdrive/MyDrive/Classical_Music_Transformer_Integration/data/midi_source/我不願讓你一個人.mid'
```

```python
midi_obj = miditoolkit.midi.parser.MidiFile(midi_path)
```

```python
lst = []

for t in midi_obj.tempo_changes:
    lst.append((t.time, t.tempo))

lst.append((midi_obj.max_tick, -1))
```

```python
lst
```

```python
table = {}

for idx in range(0, len(lst)-1):
    pair = lst[idx]
    time = pair[0]
    tempo = pair[1]

    table[tempo] = lst[idx+1][0] - time

main_tempo = -1
max_ticks = 0
for key, value in table.items():
    if value > max_ticks:
        max_ticks = value
        main_tempo = key
```

```python
main_tempo
```

```python
fctr = main_tempo / 60 # scale (in this case stretch) the overall tempo by this factor
score = music21.converter.Converter()
score.parseFile(midi_path)
newscore = score.stream.augmentOrDiminish(fctr)
newscore.write('midi','/gdrive/MyDrive/Classical_Music_Transformer_Integration/data/midi_source/' + midi_path.split('/')[-1])
```
